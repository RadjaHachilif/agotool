```python
import pandas as pd
import numpy as np
import tools
```

```python
pd.set_option("max_columns", 500)
```

```python
fn = r"/Volumes/JVO/RJC/ancient/20150501 459-612 txt/evidence.txt"
dfe = pd.read_csv(fn, sep='\t')
```

```python
dfe = tools.remove_reverse_rows(dfe)
```

```python
an_list = ["P02754", "P67976", "P02756", "P02755", "P00711", "P02662", "P02666"]
```

```python
def part_is_in(string_, an_list):
    if type(string_) == float and np.isnan(string_):
        return False
    for an in an_list:
        if an in string_:
            return True
    return False
```

```python
cond = dfe["Proteins"].apply(part_is_in, args=(an_list, ))
print sum(cond)
indices_2_keep = dfe[cond].index.tolist()
cond_notcont = pd.isnull(dfe["Potential contaminant"])
indices_2_keep += dfe[cond_notcont].index.tolist()
dfe = dfe.loc[indices_2_keep, :] 
```

```python
# group by RawFile
# remove duplicates in sequence col 
# count number of peptides
df2write = dfe.groupby("Raw file")["Sequence"].apply(lambda s: s.drop_duplicates().count())
df2write = df2write.reset_index()
```

```python
df2write.head()
```

```python
protein_group_ids = tools.commaSepCol2uniqueFlatList(dfe, "Protein group IDs")
protein_group_ids = [int(ele) for ele in protein_group_ids]
```

```python
import transform
```

```python
# before applying filter criteria from evidence file
# quantile normalization
fn_pg = r"/Volumes/JVO/RJC/ancient/20150501 459-612 txt/ISBA/proteinGroups.txt"
fn_out = r"/Volumes/JVO/RJC/ancient/20150501 459-612 txt/ISBA/proteinGroups_qnorm.txt"
df = pd.read_csv(fn_pg, sep='\t')
columns_to_quantile_normalize = ['Intensity ']#, 'iBAQ '] # DON'T normalize LFQ
transform.quantile_norm_Kelstrup_preserveDFshape(fn_pg, fn_out, colname_part_list=columns_to_quantile_normalize)
df = pd.read_csv(fn_out, sep='\t')
```

```python
# apply filtering criteria from evidence file to proteinGroups
df = df.loc[protein_group_ids, :]
```

```python
cols = tools.get_colnames_containing_word(df, "Intensity ")
```

```python
for col in cols:
    cond = df[col] > 0
    protein_count = df.loc[cond, col].count()    
    col = col.replace("Intensity ", "")
    cond = df2write["Raw file"].apply(tools.contains_word, args=([col], ))
    df2write.loc[cond, "Protein_count"] = protein_count
df2write["Protein_count"] = df2write["Protein_count"].astype(int)    
```

```python
df2write.columns = ["Raw file", "Peptide_count", "Protein_count"]
```
